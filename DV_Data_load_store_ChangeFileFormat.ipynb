```python
import pandas as pd
```

```python
data ={
    "CustomerID": [101,102,103,104],
    "CustomerName": ["Shashi","yash","veman","chippa"],
    "City": ["mumbai","pune","delhi","banglore"],
    "orderAmount": [1000,2000,4000,899]
}
```

```python
df=pd.DataFrame(data)

print(df)
```

```python
#storing data in text file format

df.to_csv("sample.txt", index=False)
print("Data stored in text format")
```

```python
#reading data from text file stored

read_df = pd.read_csv("sample.txt")
print("data read from text file format :")
print(read_df)
```

```python
#saving data in other file formats :JSON, excel

#JSON file format

df.to_json("sample1.json")#, orient="records"
print("Data saved to json file format")
```

```python
#reading data from json file

read_df1= pd.read_json("sample1.json")
print("data from json file is :")
print(read_df1)
```

```python
#saving data to excel file format

df.to_excel("sample2.xlsx",index=False)
print("data saved to excel format")
```

```python
#reading data from excel file

read_df2 = pd.read_excel("sample2.xlsx")
print("data from excel file is:")
print(read_df2)
```
