```python
import pandas as pd
```

```python
data = {
    "custonerID": [101,102,103,104,105, None],
    "Name": ["meera","janabai", None , "tukaram","chokhamedha","chippa"],
    "amount": ["2500","1800", None, "2222",None,"abc"],
    "city":[ "mumbai", "pune", "delhi", None, "delhi","lucknow"]
}
```

```python
df = pd.DataFrame(data)
```

```python
print (df.head(5))
```

```python
#checking we have missing values

df.isnull().sum()
```

```python
df.info() # checking the data type of our DataFrame
```

```python
df.describe() #checking statistical summary
```

```python
df.shape #checking shape of data
```

```python
#handling missing values

df["custonerID"]=df["custonerID"].fillna(df["custonerID"].mode())
```

```python
df
```

```python
#if name of customer or city is missing
df["Name"] = df["Name"].fillna("Unknown")
df["city"] = df["city"].fillna("Unknown")
df["amount"] = df ["amount"].fillna("Unknown")
print(df)
```

```python
#removing any extra spaces

df["Name"] = df["Name"].str.strip().str.title()
df["city"] = df["city"].str.strip().str.title()
```

```python
df["amount"] = df["amount"].str.replace(" ","", regex = False ) #removing spaces from amount
```

```python
df["amount"] = pd.to_numeric(df["amount"], errors="coerce")
```

```python
df.info()
```

```python
df["amount"]= df["amount"].fillna(df["amount"].mean())
```

```python
df
```

```python
df.drop_duplicates(inplace= True)
```
