```python
#dropping column PassengerID
titanic.drop("PassengerId",axis=1,inplace=True)
titanic.columns
```

```python
titanic.drop("Cabin",axis=1,inplace=True)
titanic.isna().sum()
```

```python
titanic.dtypes
```

```python
titanic["Age"].skew()
```

```python
# univariate analysis
sns.kdeplot(x="Age", data=titanic)
```

```python
# as data is normally distributed, replace the missing values with mean of the column
titanic["Age"] = titanic["Age"].fillna(titanic["Age"].mean())
titanic.isna().sum()

titanic["Embarked"].value_counts()

m=titanic["Embarked"].mode().item()
titanic["Embarked"] = titanic["Embarked"].fillna(m)
titanic.isna().sum()

```

```python
#Plotting the boxplot
sns.boxplot(x='Age', data=titanic)
```
